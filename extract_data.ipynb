```python
def find_tags(tags, tag_set):
    tagged = list(set(tags).intersection(tag_set))
    if tagged:
        return tagged[-1]
    else:
        return None
```

```python
recipe_table.cuisine = recipe_table.cuisine.apply(find_tags, args=(cuisines,))
```
